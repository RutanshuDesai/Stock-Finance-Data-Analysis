## Analysis on Max volume (Max transactions)

```python
fb['Volume'].max()
# Max Transactions = 107 Million           # 28th Jan, 2016
```

```python
fb['Volume'].mean()
# Average daily transactions = 25 Milliion
```

```python
fb['Volume'].min()
# Min Transactions = 8.65 Million             # 25th Nov, 2016
```

```python
fb['Volume'].idxmax()
# the date is 28th January 2016, max trade in the year 2016
```

```python
fb['Volume'].idxmin()
```

```python
fb[['Open','Close']].mean()
# average stock price for the year 2016
```

```python
fb[['Open','Close']].max()
# maximum stock price 
```

```python
fb.loc['2016-01-28'][['Open','Close']]
# price on maximum traded day
```

```python
fb.loc['2016-11-25'][['Open','Close']]
# price on minimum trade
```

```python
trade_jan_mar = fb.ix['2016-01-01':'2016-03-01']
trade_jan_mar.head()
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
plt.figure(figsize=(10,10))
trade_jan_mar['Open'].plot()
# So an interesting thing is that just few days before the sudden stock price rise, 
# market had seen its largest volume for the year 2016. The analyst were able to predict that the prices are going to boost.
```

```python
plt.figure(figsize=(20,10))
fb ['Close'].plot()
# This line plot shows how drastically the price rose during the January end and start of February. 
# So may be the traders predicted well that there is good chnce of making money.
# For later part during the year, there isn't any such sudden spike seen as compared to Jan-Feb.
```
